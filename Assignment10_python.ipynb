5. Create a programme that searches a folder tree for files with a certain file extension (such as .pdf or .jpg). Copy these files from whatever location they are in to a new folder.


<span style='color:green'>Ans.</span>

```python
import os, shutil
def moveFileType(folder):
    for root, dirs, files in os.walk(folder):
        for file in files:
            if file.endswith('.jpg'):
                image_path=os.path.join(root,file)  # get the path location of each jpg image.
                print('location: ',image_path) 
                shutil.copy(image_path, '/Users/JijoPinks/')
                
moveFileType('/Users/JijoPinks/')

```
